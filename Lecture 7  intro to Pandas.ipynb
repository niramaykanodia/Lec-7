# Report card dictionary

```python
report_card_dictionary = { "student_names": ["Niramay","aryan", "yash", "shubham"],
                          "math_scores": [45,42,41,40],
                          "bio_scores": [50,45,30,80],
                          "history_score": [80,100,90,85],
                          "physics_scores": [80,58,90,45]
                         }
report_card_dictionary
                          
                   
```

# set up

```python
import pandas as pd
```

# covert dictionary to data frame

```python
report_card_df = pd.DataFrame(report_card_dictionary)
report_card_df

#report_card_df.head() if data is too long
```

# Add Colums called total score

```python
report_card_df["total_score"] = report_card_df["math_scores"] + report_card_df["bio_scores"] + report_card_df["history_score"] + report_card_df["physics_scores"]
report_card_df
```

# Create another colum called avg score

```python
report_card_df["average_score"] = report_card_df["total_score"]/4
report_card_df
```
